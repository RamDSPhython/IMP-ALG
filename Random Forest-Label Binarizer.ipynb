```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Reading CSV File
df = pd.read_csv("Tree.csv")
```

```python
#Calling Tree.csv file as Head 
df.head()
```

```python
#seperating Character data set from tree data set
df_car=df.select_dtypes(include=[object])
```

```python
#Calling Character data set
df_car
```

```python
#Seperating Number data set from Tree Data set
df_num = df.select_dtypes(include=[np.number])
```

```python
#Calling Number Data Set 
df_num
```

```python
#Importing Label binarizer to change the lable into 1's and 0's 
from sklearn.preprocessing import LabelBinarizer
```

```python
#Assining Label Binarizer function to lb
lb = LabelBinarizer()
```

```python

#Converting df_car column labels to binary format
df_car['company'] = lb.fit_transform(df_car['company'])
df_car['job'] = lb.fit_transform(df_car['job'])
df_car['education'] = lb.fit_transform(df_car['education'])
```

```python
#Calling Character data after label binarizer
df_car
```

```python
#Assining  X data 
x=df_car
```

```python
#Calling x data
x
```

```python

#Assining Y Data

y=df_num
```

```python
#Calling y data 
y
```

```python
#importing Model selecting libaray
from sklearn.model_selection import train_test_split
```

```python
#Assigning Test and train sizes
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
#Importing Random Forest Library
from sklearn.ensemble import RandomForestClassifier
```

```python
#Assingning it to library
RFC = RandomForestClassifier()
```

```python
#Fitting Train data 
RFC.fit(x_train,y_train)
```

```python
#Caluclating score for train data 
RFC.score(x_train,y_train)
```

```python
#Predicting Test data for Random Forest Alg
pre=RFC.predict(x_test)
```

```python
#Predicting Sample data values 
val1 =RFC.predict([[0,1,1]])
val1
```
